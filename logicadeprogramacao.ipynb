#Laços e Loops

```python
idades
```

```python
#for fora da função

def verificar_se_pode_dirigir(idade):
  if idade >= 18:
    print(f'{idade} anos de idade, TEM permissão para dirigir.')
  else:
    print(f'{idade} anos de idade NÃO TEM permissão para dirigir.')

for idade in idades: #para cada idade em nossa lista
  verificar_se_pode_dirigir(idade)
```

```python
#for dentro da função

def verificar_se_pode_dirigir(idades):
  for idade in idades:
    if idade >= 18:
      print(f'{idade} anos de idade, TEM permissão para dirigir.')
    else:
      print(f'{idade} anos de idade NÃO TEM permissão para dirigir.')

verificar_se_pode_dirigir(idades)
```

```python
permissoes = []
idades = [20, 14, 40]

def verifica_se_pode_dirigir(idade, permissoes):
  for idade in idades:
    if idade >= 18:
      #adicionar os valores true na lista permissoes, 
      #e para isso utilizaremos o appends()
      permissoes.append(True)
    else:
      permissoes.append(False)
verifica_se_pode_dirigir(idade, permissoes)
```

```python
permissoes
```

```python
#se a permissao for True, então pode dirigir
for permissao in permissoes:
  if permissao == True:
    print('Pode dirigir.')
  else:
    print('Não pode dirigir.')
```

#Exercícios

```python
import matplotlib.pyplot as plt

notas_matematica = ['Matemática',8,7,6,6,7,7,8,10]
notas_portugues = ['Português',9,9,9,8,5,6,8,5]
notas_geografia = ['Geografia',10,10,6,7,7,7,8,7]

notas = [notas_matematica, notas_portugues, notas_geografia]
```

```python
for nota in notas:
 x = list(range(1, 9)) #8 provas
 y = nota[1:] #notas a partir dos números, pula o nome das matérias na posição 0
 plt.plot(x, y, marker='o')
 plt.xlabel('Provas')
 plt.ylabel('Notas')
 plt.title(nota[0])
 # plt.ylim([valor_minimo, valor_maximo]) faz os eixos começarem do ZERO
 plt.ylim([0, 10])
 plt.show()
```
